```python
import os
import time
import sys
import random
num = '0123456789qQ'
# create classes
# main character/player class
class character:
    def __init__(self, name, HP, MP, nature, dmg, level, accuracy, exp):
        self.name = name
        self.dmg = dmg
        self.superdmg = dmg * 3
        self.nature = nature
        self.HP = HP
        self.MP = MP
        self.accuracy = accuracy
        self.level = level
        self.exp = exp
        self.HP_org = HP
        self.MP_org = MP
        self.dmg_org = dmg
    # attach function of player including basic attack and ultimate attack
    def attack(self, villain, opt):
        time.sleep(0.05)
        if opt == 0: 
            if random.random() < self.accuracy:
                villain.HP -= self.dmg
                print("Eat my furious fist", str(villain.name), "!!!")
                print("You hit", str(villain.name), "and did", str(self.dmg), "points of damage!")
            else:
                villain.HP -= 0
                print("You missed! No damage was made.")
                
        if opt == 1:
            if random.random() < self.accuracy / 1.5:
                villain.HP -= self.superdmg
                print("O-LA-O-LA-OOOOOOO--LAAAAAAAAAA!!!!!! You can eat my furious fist", str(villain.name), "!!!")
                print("You hit", str(villain.name), "with you ULT and did", str(self.superdmg), "points of damage!")
            else:
                villain.HP -= 0
                print("You missed! No damage was made.")
    # define cast magic function            
    def cast(self, villain, skill):
        time.sleep(0.05)
        self.MP -= skill.MPcost
        if  (
            (skill.nature == "water" and villain.nature == "fire") or (skill.nature == "fire" and villain.nature == "earth") or
            (skill.nature == "earth" and villain.nature == "wind") or (skill.nature == "wind" and villain.nature == "water")
            ):
            print("ARRRRHHHHH! RELEASE! My Magic Power! Destroy my enemy!")
            print("Your magic", str(skill.name), "is very effective!")
            villain.HP -= skill.skdmg * 3
            time.sleep(0.05)
            print("You casted", str(skill.name), "to", str(villain.name), "and did", str(skill.skdmg * 3), "points of damage!")
        else:
            villain.HP -= skill.skdmg       
            print("Be with me! My Magic Power! Destroy my enemy!")
            time.sleep(0.05)
            print("You casted", str(skill.name), "to", str(villain.name), "and did", str(skill.skdmg), "points of damage!")
    # define self heal function   
    def heal(self):
        time.sleep(0.25)
        if self.HP + self.dmg * 2 <= self.HP_org:
            self.HP += self.dmg * 2
            self.MP -= self.dmg / 2
            print("Heal!")
            print("You gained ", str(self.dmg * 2), " points of HP.")
        else:
            self.HP = self.HP_org
            self.MP -= self.dmg / 2
            print("Heal!")
            time.sleep(0.25)
            print("Your HP is now full!")
    # define use items function
    def use(self, item):    
        time.sleep(0.05)
        self.HP += item.HP_add
        self.MP += item.MP_add
        item.amount -= 1
        print("You used ", str(item.name), "and added ", str(item.HP_add), "points of HP.")
        print("You used ", str(item.name), "and added ", str(item.MP_add), "points of MP.")
    # define weild weapon function. A combo after attack function    
    def weild(self, weapon, villain):
        time.sleep(0.05)
        villain.HP -= weapon.dmg_add
        print("You followed up with a combo-strike using", str(weapon.name), "on", str(villain.name))
        print("You did additional", str(weapon.dmg_add), "points of damage." )
        weapon.damage()
    # define level up and corresponding ability enhancement
    def levelup(self): 
        time.sleep(0.15)
        self.level += 1
        self.HP = self.HP_org
        self.MP = self.MP_org
        self.HP += 100
        self.MP += 100
        self.HP_org = self.HP
        self.MP_org = self.MP
        self.accuracy += 0.1
        self.dmg = self.dmg_org
        self.dmg += 50
        self.dmg_org = self.dmg
        print("LEVEL UP!")
        print("Your HP is now",str(self.HP), "and MP is now", str(self.MP))
        print("Your strenth increased 50 points!")
    # define reborn if the player dies   
    def reborn(self):
        time.sleep(0.15)
        self.HP = self.HP_org
        self.MP = self.MP_org
        print("A shining dragon spirit came from the sky and blew a light breath at you......")
        time.sleep(0.18)
        print("You slowly opened you eyes......You were reborned......")
    
    # define the major battle sequences, including attack, cast, use, adn other functions    
    def battle(self, Skill_tree, Item_bag, weapon, villain):
        os.system('clear')
        print("Fight! You have encountered ", str(villain.name), "!")
        while self.HP > 0 and villain.HP > 0:
            print("")
            print("You HP is ", str(self.HP), " and MP is ", str(self.MP))
            print(str(villain.name),"'s HP is ", str(villain.HP))
            print("")
            choice = input("Select attack options: Cast press [ 1 ], Attack press [ 2 ], Heal press [ 3 ] , use items press [ 4 ]\n") or str(-999)
            if choice not in num:
                print("Enter a valid input.")
                continue
            choice = int(choice)   
            time.sleep(0.05)
            if int(choice) == 1:
                for i in range(len(Skill_tree)):
                    print("Cast >>>", str(Skill_tree[i].name), "press [", str(i), "]")
                sk = input("\n") or -999
                if sk not in num:
                    print("Enter a valid input.")
                    continue
                sk = int(sk)  
                if sk < len(Skill_tree) and sk >= 0:
                    skill = Skill_tree[sk]
                else:
                    print("input a valid option")
                    continue
                if self.MP - skill.MPcost >= 0:
                     self.cast(villain, skill)
                else:
                    print("Not enough MP.")
                    continue
                
            elif int(choice) == 2:
                print("")
                opt = input("Select attack option: Basic Attack press [ 0 ], Ult-Attack press [ 1 ]\n") or str(-999)
                if opt not in num:
                    print("Enter a valid input.")
                    continue
                opt = int(opt) 
                if opt == 0 or opt == 1:
                    self.attack(villain, opt)
                else:
                    print("Enter a valid option dude. You are in a fight.")
                    continue
                if weapon.dur > 0 and villain.HP > 0:
                    time.sleep(0.05)
                    print("You can use", str(weapon.name))
                    print("")
                    opt2 = input("Would you like to use it? Yes press [ 1 ] and No press [ 0 ]\n") or str(-999)
                    if opt2 not in num:
                        print("Enter a valid input.")
                        continue
                    opt2 = int(opt2) 
                    time.sleep(0.05)
                    if opt2 == 1:
                        if weapon.dur > 1:
                            self.weild(weapon, villain)
                        elif weapon.dur == 1:
                            self.weild(weapon, villain)
                            weapon.broken()
                        else:
                            continue
                    elif opt2 == 0:
                        pass
                    else:
                        print("Enter a valid option dude. You are in a fight.")
                        continue
                
            elif choice == 3:
                if self.MP - self.dmg / 2 >= 0:
                    self.heal()
                else:
                    print("Not enough MP.")
                    continue
                
            elif choice == 4:
                for i in range(len(Item_bag)):
                    print("Use >>>", str(Item_bag[i].name), "press [", str(i), "]    ", str(Item_bag[i].amount), "remaining")
                print("")
                it = input("\n") or str(-999)
                if it not in num:
                        print("Enter a valid input.")
                        continue
                it = int(it) 
                if it < len(Item_bag) and it >= 0:
                    item = Item_bag[it]
                else:
                    print("input a valid option")
                    continue
                if item.amount > 0:
                    self.use(item)
                else:
                    print("Not enough item.")
                    continue
        
            else:
                print("Enter a valid option dude. You are in a fight.")
                continue
            # the villain's attack round
            time.sleep(0.5)
            print("")
            if villain.HP > 0 and self.HP > 0: 
                ch = random.random()    
                if (villain.MP - villain.dmg) >= 0 and ch > 0.7:
                    villain.cast(self)
                else:  
                    optv = random.random()
                    if optv < 0.7:
                        optv = 1
                    else:
                        optv = 2
                    villain.attack(self, optv)
                time.sleep(1)
                os.system('clear')
            
        if self.HP <= 0:
            time.sleep(0.8)
            print("You died. Try again next time.")
            flag = 0
            return flag
        else:
            time.sleep(0.8)
            print("YOU WIN! You have defeated", str(villain.name))
            flag = 1
            return flag 

# define villains/bosses
class villain:
    def __init__(self, name, HP, MP, nature, dmg, level, accuracy, exp, bsk_name):
        self.name = name
        self.dmg = dmg
        self.superdmg = dmg * 3
        self.nature = nature
        self.HP = HP
        self.MP = MP
        self.accuracy = accuracy
        self.level = level
        self.exp = exp
        self.heal_add_HP = HP * 0.2
        self.bsk_name = bsk_name 
        
    # boss attack function   
    def attack(self, player, opt):    
        time.sleep(0.5)
        if opt == 1:
            if random.random() < self.accuracy:
                player.HP -= self.dmg
                print("")
                print(str(self.name), ": RARRRRRRRR! Die", str(player.name), "!!!")
                print(str(self.name) , "hit", str(player.name), "and did", str(self.dmg), "points of damage!")
            else:
                player.HP -= 0
                print(str(self.name), "missed! No damage was made.")
                
        if opt == 2:
            if random.random() < self.accuracy / 2:
                player.HP -= self.superdmg
                print("")
                print(str(self.name), ": ROARRRRR! KNEEL BEFORE ME!!!!", str(player.name), "!!!")
                time.sleep(0.5)
                print(str(self.name), "hit", str(player.name), "with ULT and did", str(self.superdmg), "points of damage!")
            else:
                player.HP -= 0
                time.sleep(0.5)
                print(str(self.name), "missed! No damage was made.")
                
    # define boss cast function         
    def cast(self, player):
        time.sleep(0.25)
        self.MP -= self.dmg
        player.HP -=  self.dmg * 2      
        print("")
        print(str(self.name), ": May The Dark Lord Grant Me Power! Destroy the Hero!")
        time.sleep(0.5)
        print(str(self.name), "casted", str(self.bsk_name), "to", str(player.name), "and did", str(self.dmg * 2), "points of damage!")
        
# define weapons player cna weild
class weapon:
    def __init__(self, name, dmg_add, dur):
        self.name = name
        self.dmg_add = dmg_add
        self.dur = dur
    # define damage of the weapon if used
    def damage(self):
        self.dur -= 1
        if self.dur <= 0:
            self.dur = 0
    # the weapon will break if durability is 0  
    def broken(self):
        self.dmg_add = 0
        print("The", str(self.name), "is broken!")
            
# define the skills a player can cast        
class skill:
    def __init__(self, name, skdmg, nature, MPcost, skaccuracy):
        self.name = name
        self.skdmg = skdmg
        self.nature = nature
        self.MPcost = MPcost
        self.skaccuracy = skaccuracy

# define items can be used        
class item:
    def __init__(self, name, HP_add, MP_add, amount):
        self.name = name
        self.HP_add = HP_add 
        self.MP_add = MP_add 
        self.amount = amount
        
# define the boss environment
class corner:
    def __init__(self, name, boss_name, description):
        self.name = name
        self.boss_name = boss_name
        self.description = description

    def corner_effect(self, player, villain):
        player.dmg *= 0.85
        player.HP *= 0.8  
        villain.dmg *= 1.15
# define a game class that takes in all the components of the game       
class game:
    def __init__(self, player, BOSS, Minions, Item_bag, Skill_tree, WEAP, broom, CORNER):
        self.player = player
        self.BOSS = BOSS
        self.Minions = Minions
        self.Item_bag = Item_bag
        self.Skill_tree = Skill_tree
        self.WEAP = WEAP
        self.broom = broom
        self.CORNER = CORNER
    # define play function of game class
    def play(self):
        while len(self.BOSS) > 0:
            print("")
            print("You must select a corner to fight a boss!", "There are", str(len(self.BOSS)), "bosses remaining.")
            print("")
            act = input("Select an action: fight minions press [ 1 ], fight bosses press [ 2 ], or press [ q ] to quit \n" ) or str(-999)
            if act not in num:
                print("Enter a valid input.")
                continue
                
            if act == 'q' or act == 'Q':
                print("See you next time!")
                sys.exit()
                
            
            act = int(act) 
            os.system('clear')
            if act == 1:
                print("You found a random minion. Give the minion some discipline!")
                flag = self.player.battle(self.Skill_tree, self.Item_bag, self.broom,  self.Minions)
                if flag == 0:
                    self.player.reborn()
                    time.sleep(2)
                
                elif flag == 1:
                    self.player.levelup()
                    self.Minions.HP = 500
                    self.Minions.MP = 0
                    time.sleep(3)   
            elif act == 2:
            
                for i in range(len(self.BOSS)):
                    print(">>>", str(self.BOSS[i].name), "press [", str(i), "]")
                time.sleep(0.25)   
                choice = input("Select a boss to fight!\n") or str(-999)
                if choice not in num:
                    print("Enter a valid input.")
                    continue
                choice = int(choice)
                time.sleep(0.5)     
                if choice < len(self.BOSS) and choice >= 0:
                    print(self.CORNER[choice].description)
                    print("You challenged", str(self.BOSS[choice].name), "and get ready!!!")
                    flag = self.player.battle(self.Skill_tree, self.Item_bag, self.WEAP[choice],  self.BOSS[choice])
                    if flag == 0:
                        self.player.reborn()
                        time.sleep(1)
                
                    elif flag == 1:
                        # 3 level up
                        self.player.levelup()
                        self.player.levelup()
                        self.player.levelup()
                        # learn new skills
                        print("The spirit of", str(self.BOSS[choice].name), "was absorbed by your dragon necklace..." )
                        time.sleep(0.5)
                        print("You have learned", str(self.BOSS[choice].bsk_name), "!!!")
                        LS = skill(str(self.BOSS[choice].bsk_name), self.BOSS[choice].dmg, self.BOSS[choice].nature, self.BOSS[choice].dmg / 2, 1) 
                        self.Skill_tree.append(LS)
                        self.BOSS.pop(choice)
                        self.CORNER.pop(choice)
                        self.WEAP.pop(choice)
                        time.sleep(1.2)
                else:
                    print("Input a valid option.")
                    continue
    
            else:
                print("You can't run away from this. The Dragon Master tought you better than this!")
                continue
    
# define main funciton, initialization of games and characters and so on
def main(): #the main funciton for the game
    os.system('clear')
    print("-------High School Gokudo -- The Dragon of Shinjuku-------")  
    print("-------------------A Wild Advanture-----------------------")  
    print("--------------------by Oliver Chang-----------------------")
    print()
    wait = input("Press [ENTER] to play.")
    os.system('clear')

    # define original skills
    WoD = skill("The Wrath of the Dragon!", 200, "fire", 50, 1) 
    WoD1 = skill("Lightning Slash of the Dragon Claws!", 300, "wind", 100, 1) 
    Skill_tree = []
    Skill_tree.append(WoD)
    Skill_tree.append(WoD1)
    # define weapons
    WEAP = ["","","",""]
    WEAP[0] = weapon("Empty Red Wine Bottle", 800, 1)
    WEAP[1] = weapon("Folding Bar Stool", 200, 10)
    WEAP[2] = weapon("Unknown Chemical Fluid in a Brown Glass Bottle", 1500, 1)
    WEAP[3] = weapon("Wooden Baseball Bat", 300, 10)
    
    broom = weapon("An Old Broom", 50, 1000000)
    
    # defin items
    Item_bag = ["","","",""]
    Item_bag[0] = item( "Sake", 500, 500, 10)
    Item_bag[1] = item( "Huge_Bento", 1200, 500, 10)
    Item_bag[2] = item( "Panda Express To-Go Box", -100000, 0, 1)
    Item_bag[3] = item( "Chocolate Bar", 1500, 500, 10)
    # The player
    Dragon_Ryu = character('Dragon Ryu', 3000, 3000, "fire", 200, 1, 0.7, 0) 
        
    # define bosses and minions
    Minions = villain('Side-kick minion punk', 500, 0, "fire", 100, 1, 0.5, 0, "none")
    BOSS=["","","",""]
    BOSS[0] = villain('Punk Baga', 3000, 2000, "earth", 300, 2, 0.5, 0, "Punk Rock Punch!")
    BOSS[1] = villain('Big Bully Yaku', 4500, 4000, "water", 400, 3, 0.5, 0, "Big Bully Slimy Smash!")
    BOSS[2] = villain('Crackhead Anigi', 5500, 2000, "fire", 400, 4, 0.6, 0, "Black Powder Boom!")
    BOSS[3] = villain('Swifty SpicyGal', 6500, 3000, "wind", 500, 4, 0.6, 0, "Pepper Tornado!")
    
    # define corners
    txt0 = '''
    You came up to the Roof Top of the tallest building in the school...
    You found a punk head dude shouting loudly with dirty words at you.
    Punk Baga: Hey man, you want to get your ass kicked hummm? Guess you don't know who I am right?
    Punk Baga: I am the Punk Baga, the lord of this roof top. Before I throw you off the roof, you better disappear NOW!
    '''
    txt1 = '''
    You heard the Cafeteria was a dangerous place so you decided to go take a look.
    You found a big bully collecting money from the students and robbing thier lunch boxes!
    Big Bully Yaku: Hey man, did you bring your money and lunch box for me? NO? You got some ___ here.
    Big Bully Yaku: Now get your money or I'll make you my lunch!!!
    '''
    txt2 = '''
    You went to the Chemistry Lab and you smelled some very bad smells.
    Crackhead Anigi: You came to buy some drug? NO? Get the hell out of my lab punk!
    Crackhead Anigi: I supplied all the drugs to the students in the school and made a great fortune,
    I will not let you destroy my empire!!!
    '''
    txt3 = '''
    You went to the Gym and figured out it had been abandoned for a long time.
    You wondered why and you found a group of goons squatting on the floor. The head of the goons shouted at you.
    Swifty SpicyGal: No one can use this gym without my permission. I AM TALKING ABOUT YOU!
    Swifty SpicyGal: I used this place to beat up students I didn't like. And YOU WILL BE NEXT!!!
    '''
    
    CORNER=["","","",""]
    CORNER[0] = corner("Roof Top", 'Punk Baga', txt0)
    CORNER[1] = corner("Cafeteria", 'Big Bully Yaku', txt1)
    CORNER[2] = corner("Chemistry Lab", 'Crackhead Anigi', txt2)
    CORNER[3] = corner("Gym", 'Swifty SpicyGal', txt3)
    
    # main story line
    intro = '''
    Dragon Ryo (You) is a hot-headed high school student, 
    who has a big heart and a strong sense of justice.
    You just finished you martial art training with High Templar - Master Dragon of the Fuji Shrine.
    Before you left, your master told you that it was your destiny to fight against evil in this world.
    '''
    print(intro)
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    
    intro2 = '''
    You just recently transferred to the worst high school in the Shinjuku area in Japan.
    The school is full of violence, disrespect, drugs, and a bunch of other issues. 
    Your mission is to become the most powerful kid and bring peace and justice to the school.
    You must defeat the minions and bosses lurking around in the campus!!!
    '''
    print(intro2)
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    time.sleep(0.5)
    print("You are now in the campus. The principle of the school begged you to restore the order of the school.")
    print("You must help him!!!")
    print("You entered the school.....\n")
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    print("A mysterious sound humming in you head........")
    print("Master Dragon: My child.....Remember the relations among all nature......" )  
    print("Water silences fire...Fire swallows earth...Earth withstands wind...and Wind morendos water...")
    print("All become one......and one becomes all......Go...My child.......\n")
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    
    # game start
    Game = game(Dragon_Ryu, BOSS, Minions, Item_bag, Skill_tree, WEAP, broom, CORNER)
    Game.play()
    
    # if winning
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    print("The dark cloud hanging over the school had disappeared")
    print("and the sun light penetrated and splashed on the ground......")
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    print("A mysterious sound hummed again......")
    print("Dragon Master: Great job my child......You have expelled the evil and restored justice......")
    print("Now.....I can finally rest...........")
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    print("Dragon Master: Oh wait. You still have to graduate from the school you know that right?")
    print("Your parents want you to go to UC Berkeley so you better get straight A's!")
    print("Now get your ass up and go to the library NOW!")
    print("Dragon Ryu: T ____ T (cry)")
    wait = input("Press [ENTER] to continue.")
    os.system('clear')
    print("I hope you enjoyed this program.")
    print("High School Gokudo -- The Dragon of Shinjuku")
    print("Presented by Oliver Chang")

main()
```
